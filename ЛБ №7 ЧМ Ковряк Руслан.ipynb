```
import numpy as np
import matplotlib.pyplot as plt

def newton_interpolation(x, y, x0):
    n = len(x)
    f = np.zeros((n, n))
    f[:, 0] = y

    for j in range(1, n):
        for i in range(n - j):
            f[i, j] = (f[i + 1, j - 1] - f[i, j - 1]) / (x[i + j] - x[i])

    ans = 0
    for j in range(n):
        prod = f[0, j]
        for i in range(j):
            prod *= (x0 - x[i])
        ans += prod

    return ans

def plot_interpolation():
    x_data = np.array([0.15, 0.16, 0.17, 0.18, 0.19, 0.20, 0.21, 0.22, 0.23, 0.24, 0.25])
    y_data = np.array([4.4817, 4.9530, 5.4739, 6.0496, 6.6859, 7.3891, 8.1662, 9.0250, 9.9742, 11.0232, 12.1825])

    x_values = [0.159, 0.234]
    y_values = [newton_interpolation(x_data, y_data, x) for x in x_values]

    xx = np.linspace(np.min(x_data), np.max(x_data), 100)
    yy = np.array([newton_interpolation(x_data, y_data, val) for val in xx])

    plt.plot(x_data, y_data, 'o', label='Дані')
    plt.plot(xx, yy, label='Багаточлен Ньютона')
    plt.plot(x_values, y_values, 'o', label='Точки')
    plt.title("Графік інтерполяційної функції Ньютона")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid()
    plt.show()

if __name__ == "__main__":
    plot_interpolation()
```
