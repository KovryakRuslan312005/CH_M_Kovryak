```
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# Задані точки
x = np.array([-2, 0, 1, 2], dtype=float)
y = np.array([-1.5, -1, -0.5, 1.5], dtype=float)

# Точки, у яких потрібно обчислити значення функції
x_tests = [-1.5, -1, -0.5, 1.5]

# Функція для обчислення інтерполяційного полінома Лагранжа та його значення у точці x_test
def lagrange_interpolation(x, y, x_test):
    # Створення об'єкту-інтерполянта
    poly = lagrange(x, y)
    # Обчислення значення поліному в точці x_test
    return poly(x_test)

# Обчислення значень інтерполяційного поліному Лагранжа у вказаних точках
for x_test in x_tests:
    f_interp = lagrange_interpolation(x, y, x_test)
    print("Значення функції у точці x =", x_test, "дорівнює", f_interp.round(3))

# Побудова графіка інтерполяційної функції
xnew = np.linspace(np.min(x), np.max(x), 100)  # Точки, за якими будуємо графік
ynew = [lagrange_interpolation(x, y, i) for i in xnew]

plt.figure(figsize=(8, 5))
plt.plot(x, y, 'ro', label='Задані точки')
plt.plot(xnew, ynew, 'b-', label='Інтерполяційний багаточлен Лагранжа')
plt.title('Інтерполяційний багаточлен Лагранжа')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```
xnew=np.linspace(np.min(x),np.max(x),100) #точки, за якими будуємо графік
ynew=[lagrange_interpolation(x,y,i) for i in xnew]

# Візуалізація інтерполяції
fig = plt.figure(figsize=(6, 3))
plt.plot(x, y, 'o', xnew, ynew)
plt.title('Lagrange Polynomial_1')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Перевірка за допомогою функції lagrange з scipy
f = lagrange(x, y)
fig = plt.figure(figsize=(6, 3))
plt.plot(xnew, f(xnew), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial_2')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
```
